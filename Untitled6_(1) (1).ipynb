```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/house_estimation/housing.csv')

# Label encode categoricals
categorical_cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating',
                    'airconditioning', 'prefarea', 'furnishingstatus']

le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

# Split features and target
X = df.drop('price', axis=1)
y = df['price']

# Normalize
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Train model
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=0)
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
import gradio as gr
import numpy as np

# Example prediction function and Gradio interface will go here...
```

```python
def predict_price(area, bedrooms, bathrooms, stories, mainroad, guestroom, basement,
                  hotwaterheating, airconditioning, parking, prefarea, furnishingstatus):

    input_data = [[area, bedrooms, bathrooms, stories, mainroad, guestroom,
                   basement, hotwaterheating, airconditioning, parking, prefarea, furnishingstatus]]

    prediction = model.predict(input_data)[0]
    return f"Estimated House Price (normalized): {round(prediction, 3)}"
```

```python
import gradio as gr

interface = gr.Interface(
    fn=predict_price,
    inputs=[
        gr.Number(label="Area (normalized)"),
        gr.Number(label="Bedrooms"),
        gr.Number(label="Bathrooms"),
        gr.Number(label="Stories"),
        gr.Radio([0, 1], label="Main Road (0=No, 1=Yes)"),
        gr.Radio([0, 1], label="Guest Room (0=No, 1=Yes)"),
        gr.Radio([0, 1], label="Basement (0=No, 1=Yes)"),
        gr.Radio([0, 1], label="Hot Water Heating"),
        gr.Radio([0, 1], label="Air Conditioning"),
        gr.Number(label="Parking Spaces (normalized)"),
        gr.Radio([0, 1], label="Preferred Area (0=No, 1=Yes)"),
        gr.Radio([0, 0.5, 1], label="Furnishing Status (0=Furnished, 0.5=Semi, 1=Unfurnished)")
    ],
    outputs="text",
    title="🏠 House Price Predictor",
    description="Predict normalized house price based on features using LASSO Forest."
)

interface.launch()
```
